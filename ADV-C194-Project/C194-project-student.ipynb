# Task 2 - Which age group have high chances of having coronary heart disease in 10 years

```python
#Read the csv
csv = pd.read_csv("framingham.csv")
#Filter and make a new dataframe for those who has chances of having coronary heart disease in 10 years
filter_in_ten_yrs = csv.loc[csv["TenYearCHD"] == 1]
filter_in_ten_yrs
```

```python
#Group by age column and count the rows of TenYearCHD column
group_by_age_Ten_Year_CHD = csv.groupby("age")["TenYearCHD"].count().reset_index()
group_by_age_Ten_Year_CHD
```

```python
#Plot a line graph to show total number of people having a chance of coronary heart disease in 10 years
label = group_by_age_Ten_Year_CHD["age"]
value = group_by_age_Ten_Year_CHD["TenYearCHD"]

plt.subplots(figsize=(19,8))

plt.plot(label, value, label="Age group and their chance of coronary heart disease in 10 years", linewidth=3.0)

plt.xlabel("Age")
plt.xticks(rotation="vertical")
plt.ylabel("TenYearCHD")
plt.title("Age group and their chance of coronary heart disease in 10 years")
plt.legend()
```
