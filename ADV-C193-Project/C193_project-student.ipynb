```python
#Task 1 
#Import all the libraries and read cardiovascular.csv
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("cardiovascular.csv")
df
```

# Average cholestrol found in various age groups

```python
#Task 2 
#Group by age and find the average cholesterol and make a dataframe out of it
group_by_age = df.groupby("age")["cholesterol"].mean().reset_index()
group_by_age
```

```python
#Task 3 
#Plot a pie graph for various age group and their cholesterol 
label = group_by_age["age"]
value = group_by_age["cholesterol"]

plt.subplots(figsize=(19,8))

plt.plot(label, value, label="Age group and their cholesterol", linewidth=3.0)

plt.xlabel("Age")
plt.xticks(rotation="vertical")
plt.ylabel("Cholesterol")
plt.title("Age group and their cholesterol")
plt.legend()
```

```python
#Task 4
#Group by age and find maximum systolic blood pressure and create a dataframe out of it
group_by_age_systolic = df.groupby("age")["systolic_blood_pressure"].max().reset_index()
group_by_age_systolic
```

```python
#Task 5
#Group by age and find maximum diastolic blood pressure and create a dataframe out of it
random_group_group = df.groupby("age")["diastolic_blood_pressure"].max().reset_index()
random_group_group
```

```python
#Task 6
#Plot a line graph to show a Correlation between systolic and diastolic blood pressure
label = random_group_group["age"]
value = random_group_group["diastolic_blood_pressure"]

plt.subplots(figsize=(19,8))

plt.plot(label, value, label="Correlation between systolic and diastolic blood pressure", linewidth=3.0)

plt.xlabel("Age")
plt.xticks(rotation="vertical")
plt.ylabel("diastolic_blood_pressure")
plt.title("Correlation between systolic and diastolic blood pressure")
plt.legend()
```
